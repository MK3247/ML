```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x,y)

def myfunc(x):
    
    return slope*x + intercept

mymodel =list(map(myfunc, x))

plt.scatter(x,y)
plt.plot(x, mymodel)
plt.xlabel("Age")
plt.ylabel("Speed")
plt.title("Age-Speed Relationship", c = 'red')
plt.show()

print(r)
print(slope)
print(intercept)
print(std_err)
print(p)
```

```python
myfunc(15)
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x,y) # r stands for relationship (-1 to 1, 0 means no relationship) 

def myfunc(x): # linear regression method i.e y = mx + c (m = slope, c = y-intercept)
    
    return slope*x + intercept #returns the corresponding y value for a given x value on the y-axis

mymodel = list(map(myfunc, x)) # list of new y values that will be used to draw the line of best fit.

plt.scatter(x,y) # scatter plot for the original data

plt.plot(x, mymodel) # line of best fit

plt.xlabel("age")
plt.ylabel("speed")
plt.title("age-speed visualization")
plt.show()
```

predicting car's speed in 10 years

```python
myfunc(5) # predicts the corresponding value of x on the y axis
```

```python
for xv, m in zip(x, mymodel):
    
    #print ('x values'+" - "+ 'new y alues')
    
    print (xv," - ", m)
```
